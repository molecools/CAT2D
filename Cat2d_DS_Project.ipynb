````python
#pca on genotype matrix plus visualisation
#pca shows us patterns or variability in a "clearer" way. It is a dimension reduction technique when you have too many observations in your data

vcf = VCF('/content/Wild_African_eggplant_SNP_dataset.vcf.gz')

genotype_matrix = []
variant_positions = []

for variant in vcf:
    row = []
    for gt in variant.genotypes:
        a1, a2, _ = gt
        if a1 == -1 or a2 == -1:
            row.append(np.nan)
        else:
            row.append(a1 + a2)
    genotype_matrix.append(row)
    variant_positions.append(f"{variant.CHROM}_{variant.POS}")

# Create matrix: SNPs x Samples
G = np.array(genotype_matrix)
G = pd.DataFrame(G, columns=samples, index=variant_positions).T

# Mean impute missing values
G_imputed = G.fillna(G.mean(axis=0))

# Run PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
pca_result = pca.fit_transform(G_imputed)

pca_df = pd.DataFrame(pca_result[:, :2], columns=["PC1", "PC2"])
pca_df["sample"] = G.index


#viz colour by sample
# mapping samples to species; rstrip removes numbers from the sample names (check what rstrip does more broadly)
species_map = {s: s.rstrip("1234567890") for s in samples}
pca_df["species"] = pca_df["sample"].map(species_map)

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.scatterplot(data=pca_df, x="PC1", y="PC2", hue="species", palette="Set2", s=60)
plt.title("PCA of Genotypes Colored by Species")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

````

[Good enough resource for PCA](https://www.geeksforgeeks.org/implementing-pca-in-python-with-scikit-learn/). Feel free to find other sources.

---
This PCA (Principal Component Analysis) plot is showing how different genotypes group together based on their species.

 - Each dot is one genotype.

 - Colors represent different species (the legend matches species names to colors).

 - The axes, PC1 and PC2, are just new "summary" variables that capture most of the variation in the data.

 - Some species are very different from others; those whose dots are in very separate parts of the plot.

 - Some species are pretty similar — their dots are all mixed together.

 - Clusters (groups of dots) mean that the genotypes within that cluster are genetically similar.

 - If two species' colors overlap a lot, their genotypes aren't that different according to the data used for the PCA.
